# **MODEL** **DEPLOYMENT**

```python
X = df.iloc[:,df.columns!='Make']
X.head()
```

```python
encoding = {"Profession":{"Salaried": 0, "Business": 1},
            "Marrital Status":{"Single": 0, "Married": 1},
            "Education":{"Graduate": 0, "Post Graduate": 1},
            "Personal loan":{"No": 0, "Yes": 1},
            "House Loan":{"No": 0, "Yes": 1},
            "Wife Working":{"No": 0, "Yes": 1}
            }
obj_df = X.replace(encoding)
obj_df.head()               
```

```python
#using KMeans algorithm
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
X_scaled = StandardScaler().fit_transform(obj_df)
X_scaled = pd.DataFrame(X_scaled,columns=['Age', 'Profession', 'Marrital Status', 'Education', 'No of Dependents',
                                          'Personal loan', 'House Loan', 'Wife Working', 'Salary', 'Wife Salary',
                                          'Total Salary','Price'])
x = X_scaled.to_numpy()
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', 
                    max_iter = 300, n_init = 10,random_state = 0)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method Plot')
plt.xlabel('Number of Clusters')
plt.ylabel('Within-Cluster Sum of Square')
plt.tight_layout()
plt.show()
```

```python
kmeans = KMeans(n_clusters = 5, init = 'k-means++', 
                    max_iter = 300, n_init = 10,random_state = 42)
kmeans.fit(X_scaled)
y = kmeans.predict(X_scaled)
y_df = pd.DataFrame(y,columns=['Class'])
final_data = pd.concat([df,y_df],axis=1) 
final_data
```

```python
sns.pairplot(final_data, x_vars = ['Age','Salary', 'Wife Salary','Total Salary','Price'], y_vars = ['Age','Salary', 'Wife Salary','Total Salary','Price'])
```

###:= Analysis based on few other factors related to vehicles:

```python
df = pd.read_csv('evdataset3.csv')
df['inr(10e3)'] = df['PriceEuro']*0.08320 
df['RapidCharge'].replace(to_replace=['No','Yes'],value=[0, 1],inplace=True)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

-> Analysis based on top speed:

```python
fig = px.bar(df,x='Brand',y = 'TopSpeed_KmH',color = 'Brand',title = 'Analysis based on top speed:',labels = {'x':'Car Brands','y':'Top Speed Km/H'})
pio.show(fig)
```

-> Cost based on top speed:

```python
df1= df[["TopSpeed_KmH", "inr(10e3)"]].groupby("TopSpeed_KmH").count()
df1=df1.sort_values("TopSpeed_KmH",ascending = False).head(10)
plt.figure(figsize=(10,7))
plt.title('Cost based on top speed')
plt.pie(x=df1["inr(10e3)"],labels=df1.index,autopct='%1.0f%%')
plt.show()
```

-> Analysis based on fastest accleration:

```python
fig = px.bar(df,x='AccelSec',y = 'Brand',color = 'Brand',title = 'Analysis based on fastest accelaration:',labels = {'x':'Accelaration','y':'Car Brands'})
pio.show(fig)
```

-> Analysis based on maximum range(in km):

```python
ax= plt.figure(figsize=(20,5))
sns.barplot(x='Brand',y='Range_Km',data=df,palette='tab10')
plt.grid(axis='y')
plt.title('Analysis based on maximum range achieved by a brand:')
plt.xlabel('Brand')
plt.ylabel('Range')
plt.xticks(rotation=45)
```

-> Cost based on maximum range:

```python
df2= df[["Range_Km", "inr(10e3)"]].groupby("Range_Km").count()
df2=df2.sort_values("Range_Km",ascending = False).head(10)
plt.figure(figsize=(10,7))
plt.title('Cost based on Maximum Range')
plt.pie(x=df2["inr(10e3)"],labels=df2.index,autopct='%1.0f%%')
plt.show()
```
